## Task 1 - Plotting
* Plot the values of occupancy percentage versus number of weeks since release, for all movies in the dataset. 
* Plot the number of other releases in the week versus occupancy percentage and weeks until the end of lifetime, for all movies.
* Plot the number of shows in the week versus occupancy percentage and weeks until the end of lifetime, for all movies.

```python
df = pd.read_csv('Data/CinemaData.csv')
# print(df.head())
plot1 = plt.figure(1)
plt.scatter(df['WeeksSinceRelease'].tolist(),df['OccPer'].tolist())
plt.show()
o_release = df['OtherReleasesInWeek'].tolist()

plot2 = plt.figure(2)
plt.scatter(o_release,df['OccPer'].tolist(),c='b')
plt.show()

plot3 = plt.figure(3)
plt.scatter(o_release,df['Lifetime'].tolist(),c='r')
plt.show()

show_week = df['ShowsInWeek'].tolist()
plot4 = plt.figure(4)
plt.scatter(show_week,df['OccPer'].tolist(),c='r')
plt.show()

plot5 = plt.figure(5)
plt.scatter(show_week,df['Lifetime'].tolist(),c='r')
plt.show()
```

## Weeks since release vs Occupancy percentage

However no movie (except one , which is mostly a re-screening after a long gap) has screened after 20weeks since release, which could mean everyone has seen it by then and the occupancy numbers are not profitable or the movie was available online ,either way no occupancy in theatres

## Other release in week vs Ocupancy percentage 
This shows that number of other releases has no predictable effect on the occupancy of a movie. The number of other releases doesnt seem to affect the choice of the movie. A possible interpretation could also be that maybe people are watching more than 1 movie in a week.

## Other release in week vs Lfetime of movie
This graph also shows that there is no competition as such between movies,mostly because people would watch a different one every week keeping the movies alive for a while.

## Shows in week vs Ocupancy percentage
This graph shows that occupancy decreases with more number of shows,which is pretty logical cause there is less demand than supply. However the difference in demand is only slightly less than supply as we can see even if there is a huge number of shows say 50+ shows , the occupancy is never less than 65% 
Movies with low occupancy despite very less shows must just be bad.
## Shows in week vs Lifetime of movie
This is another demand and supply situation where if there is less supply i.e less number of shows , the movie runs longer so that everyone can watch it, however if the supply is huge ,many can watch it at the same time and sooner so the movies lifetime is shorter
